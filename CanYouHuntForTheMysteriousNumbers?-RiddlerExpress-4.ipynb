```python
import pandas as pd

products = [1458, 128, 2688, 125]

hdict = []

for p in products:
    for i in range(10):
        for j in range(10):
            for k in range(10):
                for l in range(10):
                    product = i*j*k*l
                    # print("The product of {}*{}*{}*{} = {}".format(i, j, k, l, product))
                    if product == p:
                        print("The product of {}*{}*{}*{} = {} and product = {}".format(i, j, k, l, product, p))
                        h = {'i':i, 'j':j, 'k':k, 'l':l, 'product':product, 'p':p}
                        hdict.append(h)

hdf = pd.DataFrame(hdict, columns=['i','j','k','l','product','p'])
hdf.head()
```

```python
products = [960, 384, 630, 270]

vdict = []

for p in products:
    for a in range(10):
        for b in range(10):
            for c in range(10):
                for d in range(10):
                    product = a*b*c*d
                    # print("The product of {}*{}*{} = {}".format(i, j, k, product))
                    if product == p:
                        solution = "{}*{}*{}*{}".format(a, b, c, d)
                        print("The product of {}*{}*{}*{} = {} and product = {}".format(a, b, c, d, product, p))
                        v = {'a':a, 'b':b, 'c':c, 'd':d, 'product':product, 'p':p}
                        vdict.append(v)

vdf = pd.DataFrame(vdict, columns=['a','b','c','d','product','p'])
vdf.head()
```

```python
# Find which numbers make the vertical products evaluate to a non-integer

qdict = []

for p in products:
    for l in range(1,10):
        quotient = p/l
        if not int(quotient) == quotient and isinstance(quotient, float):
            print("The quotient of {}/{} = {} and dividend = {}".format(p, l, quotient, p))
            q = {'dividend':p, 'divisor':l, 'quotient':quotient, 'p':p}
            qdict.append(q)
            
qdf = pd.DataFrame(qdict, columns=['dividend', 'divisor','quotient','p'])
qdf.head()

# Middle column cannot have 3, 6, 9 and right-most column cannot have 4, 5, 8
```
